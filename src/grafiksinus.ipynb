```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
rho = np.zeros((20,20))
rho1d = np.zeros(20)
nz = np.arange(1,4)
nr = np.arange(1,4)
z = np.arange(0,20)
r = np.arange(0,20)
```

```python
for pz in (z):
    sin1D = np.sin(2*np.pi*pz)/20
    rho1d[pz] = sin1D
plt.plot(z, rho1d)
plt.show()
```

```python
for pz in z:
    for pr in r:
        SinFunc = (np.sin(1*np.pi*pz)/10)*(np.sin(1*np.pi*pr)/10)
        rho[pz, pr] = SinFunc
```

```python
plot_x = np.linspace(5,15,20)
plot_y = np.linspace(5,15,20)
plot_x, plot_y = np.meshgrid(plot_x, plot_y)

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(plot_x, plot_y, rho, cmap='viridis')
ax.set_xlabel('r')
ax.set_ylabel('z')
ax.set_zlabel('konsentrasi')
ax.set_title('rho')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Wave parameters
amplitude = 1.0
frequency_x = 1.0
frequency_y = 2.0
phase_offset = 0.0

# Grid dimensions
x = np.linspace(0, 4*np.pi, 100)  # X-axis values
y = np.linspace(0, 4*np.pi, 100)  # Y-axis values
X, Y = np.meshgrid(x, y)  # 2D grid of X and Y values

# Calculate Z values (sine wave)
Z = amplitude * np.sin(frequency_x * X + frequency_y * Y + phase_offset)

# Plot the 2D sine wave
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Sine Wave')
plt.show()
```
