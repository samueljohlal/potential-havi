```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
rho = np.zeros((20,20))
rho1d = np.zeros(20)
nz = np.arange(1,4)
nr = np.arange(1,4)
z = np.arange(0,20)
r = np.arange(0,20)
```

```python
for pz in (z):
    sin1D = np.sin(2*np.pi*pz)/20
    rho1d[pz] = sin1D
plt.plot(z, rho1d)
plt.show()
```

```python
for jumlah_z in nz:
    for jumlah_r in nr:
        for pz in z:
            for pr in r:
                SinFunc = (np.sin(jumlah_z*np.pi*pz)/20)*(np.sin(jumlah_r*np.pi*pr)/20)
                rho[pz, pr] = SinFunc
```

```python
plot_x = np.linspace(5,15,20)
plot_y = np.linspace(5,15,20)
plot_x, plot_y = np.meshgrid(plot_x, plot_y)

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(plot_x, plot_y, rho, cmap='viridis')
ax.set_xlabel('r')
ax.set_ylabel('z')
ax.set_zlabel('konsentrasi')
ax.set_title('rho')
```
