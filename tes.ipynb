```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random

%matplotlib inline
%matplotlib widget
```

```python
rhomult = np.loadtxt('build/rhomult.csv', delimiter=',')
rhomult
```

```python
rhomult = rhomult.reshape(len(rhomult),20,20)
```

```python
outmult = np.loadtxt('build/outmult.csv', delimiter=',')
outmult = outmult.reshape(len(outmult),20,20)
```

```python
plot_x = np.arange(20)
plot_y = np.arange(0,20)
PLOT_X, PLOT_Y = np.meshgrid(plot_x, plot_y)
```

```python
index = random.randint(0, len(rhomult))
print(index)
```

```python
fig = plt.figure(figsize=(12,6))

ax1 = fig.add_subplot(projection='3d')
# ax1 = fig.subplots(projection='3d')
ax1.plot_surface(PLOT_X, PLOT_Y, rhomult[index], cmap='viridis')
ax1.set_xlabel('r')
ax1.set_ylabel('z')
ax1.set_zlabel('konsentrasi')
ax1.set_title('rho')
```

```python
fig = plt.figure(figsize=(12,6))

ax2 = fig.add_subplot(projection='3d')
ax2.plot_surface(PLOT_X, PLOT_Y, outmult[index], cmap='viridis')
ax2.set_title('phi')
```

```python
k = np.arange(0, 20)
l = np.arange(0, 20)
K, L = np.meshgrid(k, l)

fig, axs = plt.subplots(1, 2, figsize=(10,5))

axs[0].imshow(rhomult[index])
axs[0].set_title('rho')

axs[1].imshow(outmult[index])
axs[1].set_title('phi')

cbar1 = fig.colorbar(axs[0].imshow(rhomult[index]), ax=axs[0])
cbar1.set_label("Electrical Potential")

cbar2 = fig.colorbar(axs[1].imshow(outmult[index]), ax=axs[1])
cbar2.set_label("Electrical Potential")
```
